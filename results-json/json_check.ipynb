```python
import json

# JSON 파일 읽기
with open('/data2/project/2026winter/jjh0709/Learnable_Reidual_Scaling/results-json/cifar10_depth200_all_results.json', 'r') as f:
    data = json.load(f)

# JSON 구조 확인 (어떤 형태인지 보기)
print("=== JSON 구조 확인 ===")
print(f"타입: {type(data)}")
print(f"키들: {data.keys() if isinstance(data, dict) else '리스트'}")
print()

# ===== 케이스 1: 딕셔너리 안에 여러 모델이 있는 경우 =====
# 예: {"model1": {...}, "model2": {...}, ...}
if isinstance(data, dict) and 'results' not in data:
    print("=== 여러 모델 결과 (딕셔너리) ===")
    results = []
    for model_key, model_data in data.items():
        if isinstance(model_data, dict) and 'results' in model_data:
            results.append({
                'model_name': model_data.get('model_name', model_key),
                'depth': model_data.get('depth', 'N/A'),
                'best_acc': model_data['results']['best_acc'],
                'best_epoch': model_data['results']['best_epoch']
            })
    
    for result in results:
        print(f"{result['model_name']:30s} | Depth: {result['depth']:3s} | "
              f"Best Acc: {result['best_acc']:.4f} ({result['best_acc']*100:.2f}%) | "
              f"Best Epoch: {result['best_epoch']:3d}")

# ===== 케이스 2: 리스트 안에 여러 모델이 있는 경우 =====
# 예: [{"model_name": ..., "results": {...}}, {...}, ...]
elif isinstance(data, list):
    print("=== 여러 모델 결과 (리스트) ===")
    results = []
    for model_data in data:
        results.append({
            'model_name': model_data['model_name'],
            'depth': model_data['depth'],
            'best_acc': model_data['results']['best_acc'],
            'best_epoch': model_data['results']['best_epoch']
        })
    
    for result in results:
        print(f"{result['model_name']:30s} | Depth: {result['depth']:3d} | "
              f"Best Acc: {result['best_acc']:.4f} ({result['best_acc']*100:.2f}%) | "
              f"Best Epoch: {result['best_epoch']:3d}")

# ===== 케이스 3: models 키 아래에 여러 모델이 있는 경우 =====
# 예: {"models": [{"model_name": ...}, ...]}
elif isinstance(data, dict) and 'models' in data:
    print("=== 여러 모델 결과 (models 키) ===")
    results = []
    for model_data in data['models']:
        results.append({
            'model_name': model_data['model_name'],
            'depth': model_data['depth'],
            'best_acc': model_data['results']['best_acc'],
            'best_epoch': model_data['results']['best_epoch']
        })
    
    for result in results:
        print(f"{result['model_name']:30s} | Depth: {result['depth']:3d} | "
              f"Best Acc: {result['best_acc']:.4f} ({result['best_acc']*100:.2f}%) | "
              f"Best Epoch: {result['best_epoch']:3d}")

# CSV로 저장
if 'results' in locals() and results:
    import csv
    
    with open('model_results.csv', 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=['model_name', 'depth', 'best_acc', 'best_epoch'])
        writer.writeheader()
        writer.writerows(results)
    
    print("\n✅ CSV 파일로 저장 완료: model_results.csv")
    print(f"총 {len(results)}개 모델 정보 추출")

# ===== Markdown 테이블로 출력 =====
print("\n=== Markdown 테이블 형식 ===")
print("| Model Name | Depth | Best Accuracy | Best Epoch |")
print("|------------|-------|---------------|------------|")
if 'results' in locals():
    for result in results:
        print(f"| {result['model_name']} | {result['depth']} | "
              f"{result['best_acc']:.4f} ({result['best_acc']*100:.2f}%) | {result['best_epoch']} |")
```
